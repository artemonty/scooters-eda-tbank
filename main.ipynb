### 1. –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫  
**–û–ø–∏—Å–∞–Ω–∏–µ:**  
–ü–æ–¥–∫–ª—é—á–µ–Ω—ã –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ `pandas`, `matplotlib`, `seaborn`, `numpy`.

```python
# –ò–º–ø–æ—Ä—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞ (–∑–¥–µ—Å—å –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è CSV-—Ñ–∞–π–ª; –∑–∞–º–µ–Ω–∏—Ç–µ 'scooter_rides.csv' –Ω–∞ —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏–π –ø—É—Ç—å –∫ —Ñ–∞–π–ª—É)
df = pd.read_csv('kicksharing.csv')
# –ö—Ä–∞—Ç–∫–∏–π –æ—Ç—á—ë—Ç –æ —Ä–∞–∑–º–µ—Ä–µ –∏ –∫–æ–ª–æ–Ω–∫–∞—Ö
print(f"‚úÖ –†–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞: {df.shape[0]:,} —Å—Ç—Ä–æ–∫ √ó {df.shape[1]} —Å—Ç–æ–ª–±—Ü–æ–≤\n")

print("üì¶ –ù–∞–∑–≤–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:")
print(", ".join(df.columns.tolist()))

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Ç–∞–±–ª–∏—á–Ω–æ–π —Ñ–æ—Ä–º–µ
info_df = pd.DataFrame({
    '–¢–∏–ø –¥–∞–Ω–Ω—ã—Ö': df.dtypes,
    '–ü—Ä–æ–ø—É—Å–∫–æ–≤': df.isna().sum(),
    '–ó–∞–ø–æ–ª–Ω–µ–Ω–æ': df.notna().sum()
})
display(info_df)


# –ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫
print("üìî –ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ —Ç–∞–±–ª–∏—Ü—ã:")
df.head()
```

### 2. –ê–Ω–∞–ª–∏–∑ –ø—Ä–æ–ø—É—Å–∫–æ–≤  
**–û–ø–∏—Å–∞–Ω–∏–µ:**  
–í—ã–ø–æ–ª–Ω–µ–Ω —Ä–∞—Å—á—ë—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –∫–∞–∂–¥–æ–º—É –ø—Ä–∏–∑–Ω–∞–∫—É. –í—ã–≤–µ–¥–µ–Ω—ã –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏ –¥–æ–ª–∏ –ø—Ä–æ–ø—É—Å–∫–æ–≤.

**–í—ã–≤–æ–¥:**  
- –ù–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤: `loyalty_accrual_bns_amt`, `education_level_cd`, `marital_status_cd`.  
- –≠—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Å–ª–µ–¥—É–µ—Ç —É—á–∏—Ç—ã–≤–∞—Ç—å –ø—Ä–∏ –¥–∞–ª—å–Ω–µ–π—à–µ–º –∞–Ω–∞–ª–∏–∑–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –∑–∞–º–µ–Ω–æ–π –∏–ª–∏ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º).


```python
# –ü—Ä–æ–≤–µ—Ä–∏–º –Ω–∞–ª–∏—á–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ
missing_counts = df.isnull().sum()
missing_counts = missing_counts[missing_counts > 0]  # –æ—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ —Å—Ç–æ–ª–±—Ü—ã —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º:")
print(missing_counts.sort_values(ascending=False))

# –ü–æ–∏—Å–∫ –∞–Ω–æ–º–∞–ª–∏–π (–≤—ã–±—Ä–æ—Å–æ–≤) –≤ —á–∏—Å–ª–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
# –ü—Ä–æ–≤–µ—Ä–∏–º, –µ—Å—Ç—å –ª–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–∞–º, –≥–¥–µ –∏—Ö –±—ã—Ç—å –Ω–µ –¥–æ–ª–∂–Ω–æ
numeric_cols = ['minute_cost', 'hold_amount', 'distance_km', 
                'nominal_price_rub_amt', 'loyalty_accrual_rub_amt', 
                'loyalty_accrual_bns_amt', 'age']
for col in numeric_cols:
    if (df[col] < 0).any():
        print(f"–ü—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ: –≤ —Å—Ç–æ–ª–±—Ü–µ {col} –æ–±–Ω–∞—Ä—É–∂–µ–Ω—ã –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è: {(df[col] < 0).sum()} —à—Ç.")

# –ü–æ—Å—Ç—Ä–æ–∏–º boxplot (—è—â–∏–∫ —Å —É—Å–∞–º–∏) –¥–ª—è –≤—ã—è–≤–ª–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤ –≤ –æ—Å–Ω–æ–≤–Ω—ã—Ö —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö
plt.figure(figsize=(12,4))
sns.boxplot(x=df['distance_km'])
plt.title("Boxplot –¥–∏—Å—Ç–∞–Ω—Ü–∏–∏ –ø–æ–µ–∑–¥–∫–∏ (–∫–º)")
plt.xlabel("–î–∏—Å—Ç–∞–Ω—Ü–∏—è, –∫–º")
plt.show()

plt.figure(figsize=(12,4))
sns.boxplot(x=df['nominal_price_rub_amt'])
plt.title("Boxplot —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –ø–æ–µ–∑–¥–∫–∏ (—Ä—É–±)")
plt.xlabel("–°—Ç–æ–∏–º–æ—Å—Ç—å, —Ä—É–±.")
plt.show()
```

### 3. –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–∏ –∏ —Å–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤  
**–û–ø–∏—Å–∞–Ω–∏–µ:**  
–ü–æ–ª—è —Å –¥–∞—Ç–æ–π –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω—ã –≤ —Ñ–æ—Ä–º–∞—Ç datetime. –î–æ–±–∞–≤–ª–µ–Ω—ã –ø—Ä–∏–∑–Ω–∞–∫–∏:  
- `trip_duration_minutes` ‚Äî –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø–æ–µ–∑–¥–∫–∏  
- `day_of_week`, `start_hour`, `month` ‚Äî –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö –ø–∞—Ç—Ç–µ—Ä–Ω–æ–≤

**–í—ã–≤–æ–¥:**  
–°–æ–∑–¥–∞–Ω—ã –∫–ª—é—á–µ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –¥–ª—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –ø–æ–≤–µ–¥–µ–Ω–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏ –æ—Ü–µ–Ω–∫–∏ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø–æ–µ–∑–¥–æ–∫.


```python
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å—Ç–æ–ª–±—Ü—ã —Å –¥–∞—Ç–æ–π/–≤—Ä–µ–º–µ–Ω–µ–º –∏–∑ —Å—Ç—Ä–æ–∫–æ–≤–æ–≥–æ —Ç–∏–ø–∞ –≤ —Ç–∏–ø datetime
df['book_start_dttm'] = pd.to_datetime(df['book_start_dttm'])
df['book_end_dttm'] = pd.to_datetime(df['book_end_dttm'])
df['local_book_start_dttm'] = pd.to_datetime(df['local_book_start_dttm'])

# –°–æ–∑–¥–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø–æ–µ–∑–¥–∫–∏ –≤ –º–∏–Ω—É—Ç–∞—Ö
df['trip_duration_minutes'] = (df['book_end_dttm'] - df['book_start_dttm']).dt.total_seconds() / 60

# –ü—Ä–æ–≤–µ—Ä–∏–º –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω–æ–π –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏: –µ—Å—Ç—å –ª–∏ –Ω—É–ª–µ–≤—ã–µ –∏–ª–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
invalid_durations = (df['trip_duration_minutes'] <= 0).sum()
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–ø–∏—Å–µ–π —Å –Ω—É–ª–µ–≤–æ–π –∏–ª–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å—é –ø–æ–µ–∑–¥–∫–∏:", invalid_durations)
# –ü—Ä–∏ –Ω–∞–ª–∏—á–∏–∏ —Ç–∞–∫–∏—Ö –∞–Ω–æ–º–∞–ª–∏–π –º–æ–∂–Ω–æ —Ä–µ—à–∏—Ç—å –∏—Ö —É–¥–∞–ª–∏—Ç—å –∏–ª–∏ –∏—Å–ø—Ä–∞–≤–∏—Ç—å:
#if invalid_durations > 0:
#    df = df[df['trip_duration_minutes'] > 0]

# –°–æ–∑–¥–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫ –¥–Ω—è –Ω–µ–¥–µ–ª–∏ (0 - –ü–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫, 6 - –í–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ) –Ω–∞ –æ—Å–Ω–æ–≤–µ –ª–æ–∫–∞–ª—å–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –Ω–∞—á–∞–ª–∞ –ø–æ–µ–∑–¥–∫–∏
df['day_of_week'] = df['local_book_start_dttm'].dt.dayofweek

# –°–æ–∑–¥–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫ —á–∞—Å–∞ –Ω–∞—á–∞–ª–∞ –ø–æ–µ–∑–¥–∫–∏ (0-23, –ª–æ–∫–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è)
df['start_hour'] = df['local_book_start_dttm'].dt.hour

# –°–æ–∑–¥–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫ –º–µ—Å—è—Ü–∞ (1-12)
df['month'] = df['local_book_start_dttm'].dt.month

# –ü—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ –Ω–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Å–æ–∑–¥–∞–Ω—ã –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ
print(df[['book_start_dttm', 'book_end_dttm', 'trip_duration_minutes', 'day_of_week', 'start_hour', 'month']].head(5))

# –û–ø—Ä–µ–¥–µ–ª–∏–º –ø–µ—Ä–∏–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –æ—Ö–≤–∞—Ç—ã–≤–∞—é—Ç –¥–∞–Ω–Ω—ã–µ
print("–î–∞–Ω–Ω—ã–µ –æ—Ö–≤–∞—Ç—ã–≤–∞—é—Ç –ø–µ—Ä–∏–æ–¥ —Å", df['book_start_dttm'].min().date(), "–ø–æ", df['book_start_dttm'].max().date())
```

### 4. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∏—Å—Ç–∞–Ω—Ü–∏–∏ –ø–æ–µ–∑–¥–æ–∫  
**–û–ø–∏—Å–∞–Ω–∏–µ:**  
–ü–æ—Å—Ç—Ä–æ–µ–Ω–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –ø–æ `distance_km`.

**–í—ã–≤–æ–¥:**  
- –ë–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –ø–æ–µ–∑–¥–æ–∫ —É–∫–ª–∞–¥—ã–≤–∞—é—Ç—Å—è –≤ 1‚Äì3 –∫–º.  
- –ï—Å—Ç—å —Ä–µ–¥–∫–∏–µ –ø–æ–µ–∑–¥–∫–∏ –Ω–∞ 10+ –∫–º ‚Äî –≤–æ–∑–º–æ–∂–Ω—ã–µ –≤—ã–±—Ä–æ—Å—ã –∏–ª–∏ –¥–∞–ª—å–Ω–∏–µ –ø–æ–µ–∑–¥–∫–∏ –≤ —Ü–µ–Ω—Ç—Ä–µ –≥–æ—Ä–æ–¥–∞.


```python
# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∏—Å—Ç–∞–Ω—Ü–∏–∏ –ø–æ–µ–∑–¥–æ–∫ (–∫–º) ‚Äî —É–±–µ—Ä—ë–º –≤—ã–±—Ä–æ—Å—ã
plt.figure(figsize=(6,4))
sns.histplot(df[df['distance_km'] < 20]['distance_km'], bins=30, kde=True)
plt.title("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∏—Å—Ç–∞–Ω—Ü–∏–∏ –ø–æ–µ–∑–¥–∫–∏ (<20 –∫–º)")
plt.xlabel("–î–∏—Å—Ç–∞–Ω—Ü–∏—è, –∫–º")
plt.ylabel("–ß–∞—Å—Ç–æ—Ç–∞ –ø–æ–µ–∑–¥–æ–∫")
plt.show()

# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø–æ–µ–∑–¥–æ–∫ (–º–∏–Ω) ‚Äî —É–±–µ—Ä—ë–º –ø–æ–µ–∑–¥–∫–∏ –¥–æ–ª—å—à–µ 120 –º–∏–Ω—É—Ç
plt.figure(figsize=(6,4))
sns.histplot(df[df['trip_duration_minutes'] < 120]['trip_duration_minutes'], bins=30, kde=True)
plt.title("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø–æ–µ–∑–¥–∫–∏ (<120 –º–∏–Ω)")
plt.xlabel("–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å, –º–∏–Ω")
plt.ylabel("–ß–∞—Å—Ç–æ—Ç–∞ –ø–æ–µ–∑–¥–æ–∫")
plt.show()

# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –ø–æ–µ–∑–¥–æ–∫ (—Ä—É–±) ‚Äî —É–±–µ—Ä—ë–º –ø–æ–µ–∑–¥–∫–∏ –¥–æ—Ä–æ–∂–µ 1000 —Ä—É–±
plt.figure(figsize=(6,4))
sns.histplot(df[df['nominal_price_rub_amt'] < 1000]['nominal_price_rub_amt'], bins=30, kde=True)
plt.title("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –ø–æ–µ–∑–¥–∫–∏ (<1000 —Ä—É–±.)")
plt.xlabel("–°—Ç–æ–∏–º–æ—Å—Ç—å –ø–æ–µ–∑–¥–∫–∏, —Ä—É–±.")
plt.ylabel("–ß–∞—Å—Ç–æ—Ç–∞ –ø–æ–µ–∑–¥–æ–∫")
plt.show()

# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–æ–∑—Ä–∞—Å—Ç–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ‚Äî –æ–≥—Ä–∞–Ω–∏—á–∏–º –æ—Ç 10 –¥–æ 80
plt.figure(figsize=(6,4))
sns.histplot(df[(df['age'] >= 18) & (df['age'] <= 80)]['age'], bins=20, kde=True)
plt.title("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–æ–∑—Ä–∞—Å—Ç–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π (18‚Äì80 –ª–µ—Ç)")
plt.xlabel("–í–æ–∑—Ä–∞—Å—Ç, –≥–æ–¥—ã")
plt.ylabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π")
plt.show()
```

### 5. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –ø–æ–ª—É

–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω `countplot` –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É `gender_cd`.

**–í—ã–≤–æ–¥:**  
- –ü—Ä–∏–º–µ—Ä–Ω–æ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –º–µ–∂–¥—É –º—É–∂—á–∏–Ω–∞–º–∏ –∏ –∂–µ–Ω—â–∏–Ω–∞–º–∏.

### 6. –ü–æ–≤—Ç–æ—Ä–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏  
**–û–ø–∏—Å–∞–Ω–∏–µ:**  
–í—ã—á–∏—Å–ª–µ–Ω–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–µ–∑–¥–æ–∫ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è. –ü–æ—Å—Ç—Ä–æ–µ–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ.

**–í—ã–≤–æ–¥:**  
- –û–∫–æ–ª–æ 70% –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å–æ–≤–µ—Ä—à–∏–ª–∏ —Ç–æ–ª—å–∫–æ –æ–¥–Ω—É –ø–æ–µ–∑–¥–∫—É.  
- –û—á–µ–Ω—å –º–∞–ª–∞—è –¥–æ–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ‚Äî –ª–æ—è–ª—å–Ω—ã–µ (—Å–¥–µ–ª–∞–ª–∏ 5+ –ø–æ–µ–∑–¥–æ–∫).  
- –ü—Ä–æ–¥—É–∫—Ç—É –Ω—É–∂–Ω–æ —Ä–∞–±–æ—Ç–∞—Ç—å –Ω–∞–¥ —É–¥–µ—Ä–∂–∞–Ω–∏–µ–º –∫–ª–∏–µ–Ω—Ç–æ–≤.


```python
# –†–∞—Å—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–µ–∑–¥–æ–∫ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
user_trip_counts = df.groupby('party_rk')['order_rk'].count()

# –û—Å–Ω–æ–≤–Ω—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –ø–æ —á–∏—Å–ª—É –ø–æ–µ–∑–¥–æ–∫ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
print(f"–í—Å–µ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π: {user_trip_counts.count()}")
print(f"–°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –ø–æ–µ–∑–¥–æ–∫ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è: {user_trip_counts.mean():.2f}")
print(f"–ú–µ–¥–∏–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –ø–æ–µ–∑–¥–æ–∫ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è: {user_trip_counts.median()}")
print(f"–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –ø–æ–µ–∑–¥–æ–∫ —É –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è: {user_trip_counts.max()}")

# –¢–æ–ø-5 –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø–æ–µ–∑–¥–æ–∫
print("\n–¢–æ–ø-5 –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø–æ–µ–∑–¥–æ–∫:")
print(user_trip_counts.sort_values(ascending=False).head(5))

# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ —á–∏—Å–ª—É —Å–æ–≤–µ—Ä—à—ë–Ω–Ω—ã—Ö –ø–æ–µ–∑–¥–æ–∫
ride_count_dist = user_trip_counts.value_counts().sort_index()
print("\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ —á–∏—Å–ª—É —Å–æ–≤–µ—Ä—à—ë–Ω–Ω—ã—Ö –ø–æ–µ–∑–¥–æ–∫ (1‚Äì10):")
print(ride_count_dist.head(10))

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è: —Å–≥—Ä—É–ø–ø–∏—Ä—É–µ–º –≤—Å–µ—Ö, —É –∫–æ–≥–æ >10 –ø–æ–µ–∑–¥–æ–∫, –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏—é "11+"
ride_count_dist_10 = ride_count_dist[ride_count_dist.index <= 10].copy()
others = ride_count_dist[ride_count_dist.index > 10].sum()
if others > 0:
    ride_count_dist_10.loc[11] = others  # –∫–∞—Ç–µ–≥–æ—Ä–∏—è "11+"

plt.figure(figsize=(7,4))
sns.barplot(x=ride_count_dist_10.index.astype(str), y=ride_count_dist_10.values)
plt.title("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ —á–∏—Å–ª—É –ø–æ–µ–∑–¥–æ–∫")
plt.xlabel("–ß–∏—Å–ª–æ –ø–æ–µ–∑–¥–æ–∫")
plt.ylabel("–ß–∏—Å–ª–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π")
plt.show()
```

–ü–æ—Å—Ç—Ä–æ–µ–Ω –≥—Ä–∞—Ñ–∏–∫ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–æ –≤—Ä–µ–º–µ–Ω–∏ —Å—É—Ç–æ–∫ (`start_hour`).
